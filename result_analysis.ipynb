```python
import pandas as pd
import numpy as np
import os
if os.getcwd().split('/')[-1] == 'Workshop-simulator':
    os.chdir('..')
from main import main, input_ranges, all_need, base_need

result = pd.read_csv('results/maeda/result.csv', index_col=0)
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
result
```

```python
profit_list = []
benefit_list = []

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)
x = []
y = []
rectangles = []
centroids = []

for i, row in result.iterrows():

    input_values = {k: v for k, v in zip(input_ranges.keys(), row[2:])}
    p, _, b = main(input_values, input_ranges,all_need, base_need, mode='A')
    x+= [p[0], p[0], p[1], p[1]]
    y += [b[0], b[1], b[0], b[1]]
    color = 'blue' if row[-1]=='Yes' else 'red'
    rectangle = patches.Rectangle((p[0], b[0]), p[1]-p[0], b[1]-b[0], linewidth=1, edgecolor=color, facecolor=color, alpha=0.3)
    rectangles.append(rectangle)
    centroids.append((np.mean(p), np.mean(b)))
    
ax.scatter(x, y, marker='.', color='black')
for rectangle in rectangles:
    ax.add_patch(rectangle)

for i in range(len(centroids)-1):
    plt.arrow(centroids[i][0], centroids[i][1], centroids[i+1][0] - centroids[i][0], centroids[i+1][1] - centroids[i][1], 
              length_includes_head=True, width = 30_0000, fc='k', ec='k') # change the head_width based on your actual data range
ax.set_xlabel('Industry Benefit')
ax.set_ylabel('Citizen Benefit')
```
